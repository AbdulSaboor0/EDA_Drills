```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.drop(["Product_Category_2","Product_Category_3"], axis =1 , inplace = True)

df.isnull().sum()
```

```python
df.info()
```

```python
sns.set_theme()
plt.figure(figsize=(5,7))
ax = sns.countplot(x = "Gender", y = None,stat='count' ,data = df, color = "green", width= 0.5)
plt.title("Male Vs Female")
plt.xlabel= "Gender"
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
sns.countplot(x='Age', data = df, hue = "Gender",color= "orange")
plt.title(  "Male Vs Female by Age group")
plt.xlabel= "Age Group"
```

```python
df.info()
```

```python
plt.figure(figsize=(10,7))
ay = sns.barplot(x = "Age",  y = "Purchase", data = df)
plt.title("Purchase by Age group")
for bars in ay.containers:
  ay.bar_label(bars)
```

```python
sns.barplot(x= "Purchase", y = "Gender", data = df, color ="violet")
plt.title("Purchase by Gender")
```
