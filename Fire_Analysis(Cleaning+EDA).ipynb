## **Exploratory Data Analysis**

---



### **Relationship between Two Numerical Variables**
**Scatter Plot:** Ideal for showing the relationship between two continuous variables. Look for patterns, trends, outliers, and correlations.

**Line Plot:** Suitable for time series data or to visualize trends over a continuous variable.


---



### **Relationship between a Numerical and Categorical Variable**

**Box Plot:** Compares the distribution of a numerical variable across different categories. Helps identify differences in central tendency, spread, and outliers.

**Violin Plot:** Similar to box plots but shows the kernel density estimation, providing more detailed information about the distribution.



---
### **Relationship between Two Categorical Variables**

**Count Plot:** Shows the frequency of occurrences for each category.

**Grouped Bar Plot:** Compares the counts of two categorical variables.

**Heatmap:** Displays the relationship between two categorical variables as a color-encoded matrix.


---
### **Exploring Multiple Variables**
**Pair Plot:** Shows scatter plots for all numerical variables in a dataset. Useful for identifying potential relationships and correlations.
